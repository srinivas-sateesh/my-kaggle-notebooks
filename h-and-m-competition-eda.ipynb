```python
transactions_train_df_obj = transactions_train.select_dtypes(include=['object']).copy()
transactions_train_df_obj.describe()
```

```python
def mem_usage(pandas_obj):
    if isinstance(pandas_obj,pd.DataFrame):
        usage_b = pandas_obj.memory_usage(deep=True).sum()
    else:
        usage_b = pandas_obj.memory_usage(deep=True)
    usage_mb = usage_b/1024 ** 2
    return "{:03.2f}MB".format(usage_mb)
```

```python
customer_id= transactions_train.customer_id
print(customer_id.head())
customer_id_cat = customer_id.astype('category')
print(customer_id_cat.head())
```

```python
print(mem_usage(customer_id))
print(mem_usage(customer_id_cat))
```

```python
transactions_train_df_converted_obj = pd.DataFrame()
for col in transactions_train_df_obj.columns:
    num_unique_values = len(transactions_train_df_obj[col].unique())
    num_total_values = len(transactions_train_df_obj[col])
    if num_unique_values/num_total_values < 0.5:
        transactions_train_df_converted_obj.loc[:,col] = transactions_train_df_obj[col].astype('category')
    else:
        transactions_train_df_converted_obj.loc[:,col] = transactions_train_df_obj[col]
print(mem_usage(transactions_train_df_obj))
print(mem_usage(transactions_train_df_converted_obj))
#compare_obj = pd.concat([transactions_train_df_obj.dtypes, transactions_train_df_converted_obj.dtypes],axis=1)
#compare_obj = ['Before','After']
#compare_obj.apply(pd.Series.value_counts)
```

```python
transactions_train[transactions_train_df_converted_obj.columns] = transactions_train_df_converted_obj
mem_usage(transactions_train)
```
